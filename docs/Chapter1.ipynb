## Single Qubits

A quantum computer processes information using what are termed as quantum bits, often abbreviated as qubits. The most basic unit of quantum information is referred to as a single qubit or simply 1 qubit. It is a complex two-dimensional unit state vector that describes a two-level quantum system:


$ \ket{\psi} = a\ket{\varphi_1} b\ket{\varphi_2} $

The vectors $a\ket{\varphi_1}$ and $b\ket{\varphi_2}$ are two arbitrary orthonormal vectors spanning the Hilbert space $ \mathcal{H} \cong  \mathbb{C}^2 $ where the qubit $ \ket{\psi} $ is, and where a and b are complex integers. $ |a|^2 $ and $ |b|^2 $ are the probabilities of the measuring the system in the state of $ \ket{\varphi_1}$ and $\ket{\varphi_2}$ respectively. That is also means $ |a|^2 +|b|^2 =1 $. $\ket{\varphi_1}$ and $\ket{\varphi_2}$ are selected orthonormal vectors since we want them distinct, that means $\braket{\varphi_r|\varphi_s} \delta_{rs} $ where r,s= 1,2.
This shows that orthonormal basis $ \varphi_{r/s} $ can be correspond to the binary values of 0 and 1. This approach lets us replicate classical computation system. Inordinary to quantum mechanics state vectors expressed as below,

$ \ket{q} = \alpha_0\ket{0} +\alpha_1\ket{1} $

The orthonormal basis of $\ket{0}$ & $\ket{1}$ are also called computational basis. Also another set of orthonormal basis exist in Hilbert space of a single qubit that are called Hadamard basis that are $\ket{+}$ & $\ket{-}$. Hadamart basis in the terms of computational basis,


$ \ket{+} = \frac{1}{\sqrt{2}} \left( \ket{0} + \ket{1} \right)  \newline  \ket{-} = \frac{1}{\sqrt{2}} \left( \ket{0} - \ket{1} \right) $

Thus computational basis in terms of Hadamard basis,

$ \ket{0} = \frac{1}{\sqrt{2}} \left( \ket{+} + \ket{-} \right) \newline \ket{1} = \frac{1}{\sqrt{2}} \left( \ket{+} - \ket{-} \right) $

Let's rewrite $ \ket{q} = \alpha_0\ket{0} +\alpha_1\ket{1} $ equation with $ \ket{+}$ and  $\ket{-}$

$ \ket{q}_H = \alpha_+\ket{+} +\alpha_-\ket{-} $ where, $\newline$ $\newline$
$\alpha_+ = \frac{\alpha_0\ket{0}+\alpha_1\ket{1}}{\sqrt{2}} ,\alpha_- = \frac{\alpha_0\ket{0}-\alpha_1\ket{1}}{\sqrt{2}}$

Since $ |\alpha_0|^2 +|\alpha_1|^2 =1 $ and they differ by a global phase, variables can be rewritten as a sphere as follows,

$ |\alpha_0| = cos(\theta/2) ,  |\alpha_1| = sin(\theta/2)$, with $\theta \in [0,\pi] \newline$
and $\ket{q(\alpha_0,\alpha_1)} $ becomes $ \ket{  q(\theta,\phi) } $ as $\newline$
$ \ket{  q(\theta,\phi) } = cos(\theta/2)\ket{0} + e^{i\phi}sin(\theta/2)\ket{1}  $, with $ \phi \in [0,2\pi] $

This equation satifies  $ |\alpha_0|^2 +|\alpha_1|^2 =1 $ and we can get demanded orthogonal and Hadamard basis as follows

$ \ket{  q(0,0) } = \ket{0} \newline
\ket{  q(\pi,0) } = \ket{1} \newline
\ket{  q(\pi/2,0) } = \ket{+} \newline
\ket{  q(\pi/2,\pi) } = \ket{-} \newline$

This sphereciation process helps visualization of the state vector of a qubit. This unit sphere called Bloch sphere. 

```python
import matplotlib.pyplot as plt
from qutip import Bloch, basis
import numpy as np

# Initialize the Bloch sphere
b = Bloch()
b.xlabel = ['$\\left|0\\right>+\\left|1\\right>$', '$\\left|0\\right>-\\left|1\\right>$']
b.ylabel = ['$\\left|0\\right>+i\\left|1\\right>$', '$\\left|0\\right>-i\\left|1\\right>$']
b.title = 'Bloch Sphere Representation'

# Define qubit state |q> = alpha_0|0> + alpha_1|1>
alpha_0 = 1 / np.sqrt(2)  # example value
alpha_1 = 1 / np.sqrt(2)  # example value

# Convert to spherical coordinates
theta = 2 * np.arccos(abs(alpha_0))
phi = np.angle(alpha_1)

# Convert to Cartesian coordinates
x = np.sin(theta) * np.cos(phi)
y = np.sin(theta) * np.sin(phi)
z = np.cos(theta)

# Add state to the Bloch sphere
b.add_vectors([x, y, z])

# Show Bloch sphere
b.show()
```

## Multible Orbits

When considering multiple qubits or n qubits (where n  2), they are defined as quantum systems with a state vector $ \ket{q_n} $ that exists within a Hilbert space. This Hilbert space is the tensor product of the Hilbert spaces of individual n qubits and is represented as $ \mathbb{C}^{2^n} $
